# Visualisation des données

```python
sns.set_palette("Paired")
sns.countplot(x='class',data=df)
```

→ Le graphe ci-dessus démontre le nombre d'occurrence par classe. C'est à dire que les personnes optent plus pour une classe économique.

```python
#countplot: permet d'afficher sous forme d'histogramme le nombre d'occurrence d'un objet selon une catégorie
sns.countplot(x='destination_city',data=df)
```

→ Le graphe ci-dessus démontre le nombre d'occurrence par destination. Selon ce dataset la destination la plus fréquentée est Mumbai

```python
#Limite le résultat aux types de données fournis
df.describe(include='object')
```

```python
#catplot: Démontre la relation entre une variable numérique et une ou plusieurs variables catégorielles
#x,y,hue : noms des variables
#kind: Correspond au type du tracé
#palette: les différentes couleurs à utiliser pour "hue"
sns.catplot(x='price',y='airline',kind='bar',hue='days_left',palette='rocket',data=df)
```

→ On remarque d'après cette représentation graphique que le prix est affecté par les jours restants avant le départ. Plus on se rapproche du jour J plus le prix est élevé.

```python
sns.catplot(y='price',x='airline',kind='bar',hue='class',palette='rocket',data=df)
```

→ On remarque que seulement deux compagnies aériennes disposent de la classe "Business" qui est clairement plus chère que la classe "économique". Aussi, ce graphique démontre la variété des prix des différentes compagnies.

```python
plt.figure(figsize=(10, 8))
plt1 = df.airline.value_counts().plot(kind='bar')
plt.title('Airline histogram', fontsize=20)
plt1.set(xlabel = 'airline', ylabel='Frequency of airline')
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.title('Source histogram')
plt1 = df['source_city'].value_counts().plot(kind='bar')
plt1.set(xlabel = 'Source city', ylabel='Frequency of source city')

plt.subplot(1,2,2)
plt.title('Destination histogram')
plt1 = df['destination_city'].value_counts().plot(kind='bar')
plt1.set(xlabel = 'Destination city', ylabel='Frequency of destination city')
plt.show()
```

→ La plupart des vols décolent de "Delhi" et atterrissent à "Mumbai"

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.title('Departure time histogram')
plt1 = df.departure_time.value_counts().plot(kind='bar')
plt1.set(xlabel = 'Departure time', ylabel='Frequency of Departure time')

plt.subplot(1,2,2)
plt.title('Arrival time histogram')
plt1 = df.arrival_time.value_counts().plot(kind='bar')
plt1.set(xlabel = 'Arrival time', ylabel='Frequency of Arrival time')
plt.show()
```

  → Au départ la plupart des vols lors du départ sont dans la matinée ou très tôt le matin.


→ A l'arrivée la plupart des vols arrivent le soir ou l'après-midi.
